```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[1, 3],
              [4, 10]])

y = np.array([5, 6])

m = np.array([-1.0, 2.0])
b = 1.0

alpha = 0.01
```

```python
def predict(X, m, b):
    return X @ m + b
```

```python
m_history = []
b_history = []
loss_history = []
```

```python
for i in range(4):

    y_pred = predict(X, m, b)
    error = y_pred - y
    loss = np.mean(error**2)

    dm = (2/len(y)) * (X.T @ error)
    db = (2/len(y)) * np.sum(error)

    m = m - alpha * dm
    b = b - alpha * db

    m_history.append(m.copy())
    b_history.append(b)
    loss_history.append(loss)

    print("Iteration:", i)
    print("m:", m)
    print("b:", b)
    print("loss:", loss)
    print("------")
```

```python
m_history = np.array(m_history)

plt.plot(m_history[:,0], label="m1")
plt.plot(m_history[:,1], label="m2")
plt.plot(b_history, label="b")
plt.legend()
plt.title("Parameter changes")
plt.show()
```

```python
plt.plot(loss_history)
plt.title("Loss over time")
plt.show()
```
